```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv(
    "28.dat", skiprows=[1], sep="\t", escapechar="#", index_col=[1, 0]
)  # [0] for 1D and [0,1] for 2D data sets
data_xr = data.to_xarray()
print(data_xr)
plt.figure(figsize=(5, 2.7))
data_xr["lockin_6_CondX"].plot(cmap="inferno")
plt.title("Sample 2,   (10x10) $\mu m^2$")
plt.show()
```
