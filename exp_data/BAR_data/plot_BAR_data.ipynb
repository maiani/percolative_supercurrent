```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv(
    "197.dat", skiprows=[1], sep="\t", escapechar="#", index_col=[0]
)  # [0] for 1D and [0,1] for 2D data sets
data_xr = data.to_xarray()
data2 = pd.read_csv(
    "202.dat", skiprows=[1], sep="\t", escapechar="#", index_col=[0]
)  # [0] for 1D and [0,1] for 2D data sets
data_xr2 = data2.to_xarray()
print(data_xr, "\n\n", data_xr2)
plt.figure(figsize=(5, 2.7))
data_xr2["lockin_1_ResX"].plot()
data_xr["lockin_1_ResX"].plot()
data_xr["lockin_2_ResX"].plot()
data_xr["lockin_3_ResX"].plot()
data_xr2["lockin_3_ResX"].plot()
plt.title("Sample 1")
plt.legend(["2 $\mu$m", "5 $\mu$m", "10 $\mu$m", "40 $\mu$m", "40 $\mu$m"])
plt.show()
```

```python
data = pd.read_csv(
    "69.dat", skiprows=[1], sep="\t", escapechar="#", index_col=[0]
)  # [0] for 1D and [0,1] for 2D data sets
data_xr = data.to_xarray()
print(data_xr)
plt.figure(figsize=(5, 2.7))
(
    data_xr["lockin_1_ResX"] - (data_xr["lockin_1_ResX"][197:202]).mean()
).plot()  # subtracted mean resistance between +-2.5 mT.
(
    data_xr["lockin_2_ResX"] - (data_xr["lockin_2_ResX"][197:202]).mean()
).plot()  # subtracted mean resistance between +-2.5 mT.
(
    data_xr["lockin_3_ResX"] - (data_xr["lockin_3_ResX"][197:202]).mean()
).plot()  # subtracted mean resistance between +-2.5 mT.
(
    data_xr["lockin_4_ResX"] - (data_xr["lockin_4_ResX"][197:202]).mean()
).plot()  # subtracted mean resistance between +-2.5 mT.
(
    data_xr["lockin_5_ResX"] - (data_xr["lockin_5_ResX"][197:202]).mean()
).plot()  # subtracted mean resistance between +-2.5 mT.
plt.title("Sample 2")
plt.legend(["2 $\mu$m", "5 $\mu$m", "10 $\mu$m", "40 $\mu$m", "40 $\mu$m"])
plt.show()
```
